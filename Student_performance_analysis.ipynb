```python
import matplotlib.pyplot as plt
import seaborn as sns

# 🎨 Set a colorful theme
sns.set(style="whitegrid", palette="bright")

# 🎯 Select only numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 🖼️ Create subplots for all numerical columns
fig, axes = plt.subplots(nrows=(len(num_cols) // 3) + 1,
                         ncols=3,
                         figsize=(15, 5 * ((len(num_cols) // 3) + 1)))

# Flatten axes so we can loop through them easily
axes = axes.flatten()

# 📊 Plot each numerical column as a colorful boxplot
for i, col in enumerate(num_cols):
    sns.boxplot(y=df[col],
                ax=axes[i],
                color=sns.color_palette("bright")[i % len(sns.color_palette("bright"))])
    axes[i].set_title(col, fontsize=12, color="black")

# ❌ Remove any extra empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# ✨ Make sure plots don’t overlap
plt.tight_layout()

# 👀 Show the plots
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Make a copy of the dataframe to keep original
df_capped = df.copy()

# Apply IQR capping and flooring only to numeric columns
numeric_cols = df_capped.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    Q1 = df_capped[col].quantile(0.25)
    Q3 = df_capped[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping and flooring
    df_capped[col] = df_capped[col].clip(lower=lower_bound, upper=upper_bound)

# ---------------- Plotting ---------------- #

# Function to make colorful boxplots
def plot_boxplots(original, capped, title):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    sns.boxplot(data=original, palette="Set2", ax=axes[0])
    axes[0].set_title(f"Before {title}", fontsize=14)

    sns.boxplot(data=capped, palette="Set3", ax=axes[1])
    axes[1].set_title(f"After {title}", fontsize=14)

    plt.show()

plot_boxplots(df, df_capped, "Outlier Treatment (IQR)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 🎯 Select only numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 🖼️ Create subplots (3 per row)
fig, axes = plt.subplots(nrows=(len(num_cols) // 3) + 1,
                         ncols=3,
                         figsize=(15, 5 * ((len(num_cols) // 3) + 1)))

# Flatten axes for easy iteration
axes = axes.flatten()

# 📊 Plot histograms with KDE (smooth curve) for each numerical column
for i, col in enumerate(num_cols):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i],
                 color=sns.color_palette("bright")[i % len(sns.color_palette("bright"))])
    axes[i].set_title(f"Distribution of {col}", fontsize=12)

# ❌ Remove empty subplots (if any)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 🎯 Select only numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 📝 Calculate skewness before transformation
skewness_before = df[num_cols].skew()
print("📊 Skewness before fixing:\n", skewness_before)

# 🛠️ Create a copy for transformed data
df_transformed = df.copy()

# 🚀 Fix skewness for every numerical column
for col in num_cols:
    skew_val = skewness_before[col]

    if skew_val > 0:
        # ✅ Positive skew → use log1p (works even if values include 0)
        df_transformed[col] = np.log1p(df[col])
        print(f"🔹 Column '{col}' was positively skewed ({skew_val:.2f}), fixed with LOG transform.")

    elif skew_val < 0:
        # ✅ Negative skew → use square (power 2)
        df_transformed[col] = np.power(df[col], 2)

for col in num_cols:
    plt.figure(figsize=(7,4))  # set figure size
    sns.histplot(df_transformed[col], bins=30, kde=True, color="green")
    plt.title(f"Distribution of {col} after skewness correction")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()  # display the plot
```

```python
from sklearn.preprocessing import LabelEncoder

# ---------------------------------------
# 📌 Step 3: Create a dictionary to store mappings
# ---------------------------------------
encoding_mappings = {}  # This will store the mapping for each categorical column

# ---------------------------------------
# 📌 Step 4: Apply label encoding to all categorical columns
# ---------------------------------------
for col in cat_cols:
    encoder = LabelEncoder()  # Create a fresh encoder for each column
    new_col_name = f"{col}_encoded"  # New column name with "_encoded"

    # Encode the categorical values and save them into the new column
    df[new_col_name] = encoder.fit_transform(df[col])

    # Store mapping: number -> original category
    encoding_mappings[new_col_name] = dict(zip(encoder.transform(encoder.classes_), encoder.classes_))

# ---------------------------------------
# 📌 Step 5: Print mappings in markdown format
# ---------------------------------------
print("### Label Encoding Mappings\n")
for col, mapping in encoding_mappings.items():
    print(f"**{col}**")  # Column name in bold
    for k, v in mapping.items():
        print(f"- {k} → {v}")  # Markdown bullet list (number → category)
    print("\n")  # Add space between columns
```

```python
# 3. Write a sample Streamlit app (save as app.py)
with open("app.py", "w") as f:
    f.write("""
# 📌 Save this as app.py and run with: streamlit run app.py
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ================================
# 🔹 Load Your DataFrame
# ================================
# Example: load your dataset
# df = pd.read_csv("your_dataset.csv")

# For demo, let's create something similar
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    "StudyTime": np.random.randint(1, 10, 50),
    "Absences": np.random.randint(0, 20, 50),
    "Grade1": np.random.randint(40, 100, 50),
    "Grade2": np.random.randint(40, 100, 50),
    "FinalGrade": np.random.randint(40, 100, 50)
})

# ================================
# 🔹 Streamlit Dashboard
# ================================
st.set_page_config(page_title="Correlation Analysis", layout="wide")
st.title("📊 Professional Correlation Analysis Dashboard")

# ---- Full Correlation Matrix ----
st.subheader("🔹 Full Correlation Matrix (Heatmap)")
corr = df.corr()

fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(
    corr, annot=True, fmt=".2f",
    cmap="coolwarm", center=0,
    cbar_kws={"shrink": 0.8},
    linewidths=0.5, ax=ax
)
ax.set_title("Correlation Matrix Heatmap", fontsize=16, weight="bold")
st.pyplot(fig)

# ---- Most Correlated Pairs ----
st.subheader("🔹 Top Correlated Column Pairs")

corr_unstacked = (
    corr.abs()
    .unstack()
    .reset_index()
    .rename(columns={"level_0": "Feature 1", "level_1": "Feature 2", 0: "Correlation"})
)

# Remove duplicates & self-correlations
corr_pairs = corr_unstacked[corr_unstacked["Feature 1"] < corr_unstacked["Feature 2"]]
top_corr = corr_pairs.sort_values(by="Correlation", ascending=False).head(10)

# Show as a neat dataframe
st.dataframe(top_corr.style.background_gradient(cmap="YlOrRd"))

# ---- Bar Chart of Correlations ----
st.subheader("🔹 Strength of Top Correlations")
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(
    x="Correlation", y="Feature 1",
    hue="Feature 2", dodge=False,
    palette="viridis", data=top_corr, ax=ax
)
ax.set_title("Top Correlated Pairs", fontsize=14, weight="bold")
st.pyplot(fig)

# ---- Scatterplots of Top Pairs ----
st.subheader("🔹 Relationship Plots of Top Correlated Pairs")
for _, row in top_corr.iterrows():
    f1, f2, corr_val = row["Feature 1"], row["Feature 2"], row["Correlation"]
    st.markdown(f"**{f1} vs {f2}** (Correlation = {corr_val:.2f})")

    fig, ax = plt.subplots(figsize=(6, 4))
    sns.scatterplot(x=df[f1], y=df[f2], ax=ax, color="teal", s=70, alpha=0.7, edgecolor="black")
    ax.set_title(f"{f1} vs {f2}", fontsize=12, weight="bold")
    st.pyplot(fig)

""")
```
